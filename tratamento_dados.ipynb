```python
import pandas as pd
import re

data = pd.read_csv('./movies_metadata.csv')
genresData= {}
releasedateData = {}
runtimeData = {}
countryData = {}
```

```python
def Merge(dict1, dict2):
    res = {**dict1, **dict2}
    return res
```

```python
regex = r'[A-Z]{1}[a-z]+'
for genre in range(len(data['genres'])):
    genresData[data['title'][genre]] = re.findall(regex,data['genres'][genre])
```

```python
regex = r'\d{4}'
for count in range(len(data)):
    releasedateData[data['title'][count]] = re.findall(regex,str(data['release_date'][count]))
```

```python
for count in range(len(data)):
    try:
        runtimeData[data['title'][count]] = int(data['runtime'][count])
    except:
        runtimeData[data['title'][count]] = data['runtime'][count]
```

```python
regex = r'[A-Z]{2}'
for count in range(len(data)):
    countryData[data['title'][count]] = re.findall(regex,str(data['production_countries'][count]))
```
